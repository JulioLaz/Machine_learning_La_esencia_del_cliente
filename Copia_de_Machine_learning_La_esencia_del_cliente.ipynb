##Funcion para graficar:

```python
import matplotlib.pyplot as plt
import seaborn as sns

def grafica_circular_y_barras(df, variable_cuantitativa, variable_cualitatitva,palette,see_set_text):
    df_groupby_vars = df.groupby(df[variable_cualitatitva].astype(str))[variable_cuantitativa].sum().reset_index().sort_values(by=variable_cuantitativa, ascending=False)

    palette = sns.color_palette(palette, len(df_groupby_vars))

    fig, ax = plt.subplots(1, 2, figsize=(18, 6), gridspec_kw={'wspace': 0})

    sns.barplot(x=variable_cuantitativa, y=variable_cualitatitva, data=df_groupby_vars, palette=palette, orient='h', ax=ax[0])
    ax[0].set_xlabel('Ventas en millones')
    # ax[0].set_title(f'Ventas por {variable_cualitatitva}', fontsize=25, color='darkblue', fontweight='bold', pad=30)
    ax[0].margins(.15, 0.05)
    total_ventas = df_groupby_vars[variable_cuantitativa].sum()

    for index, value in enumerate(df_groupby_vars[variable_cuantitativa]):
      if see_set_text:
        ax[0].text(value + 500, index, f'${str(round(value))}mll', ha='left', va='top')
        ax[0].text(value + 500, index, f'{str(round(value*100/total_ventas,2))}%', ha='left', va='bottom')
      else:
        ax[0].text(value + 500, index, f'{str(round(value*100/total_ventas,2))}%', ha='left', va='bottom')

    # GrÃ¡fica circular
    explode = [0] * len(df_groupby_vars)
    wedges, texts, autotexts = ax[1].pie(df_groupby_vars[variable_cuantitativa], labels=df_groupby_vars[variable_cualitatitva], explode=explode, autopct='', startangle=60, wedgeprops=dict(width=0.5), colors=palette)

    total_pago_tipo = df_groupby_vars[variable_cuantitativa].sum()


    for i, (wedge, text, autotext) in enumerate(zip(wedges, texts, autotexts)):
        pago_tipo = df_groupby_vars.iloc[i][variable_cualitatitva]
        frecuencia = df_groupby_vars.iloc[i][variable_cuantitativa]
        text.set_color('black')

        if see_set_text:
          text.set_text(f'{pago_tipo}\n${round(frecuencia)}mll\n{round(frecuencia*100/total_pago_tipo,2)}%')
        else:
          text.set_text(f'{pago_tipo}\n{round(frecuencia*100/total_pago_tipo,2)}%')

        if see_set_text:
          ax[1].text(0, 0, f'Total\n${round(total_pago_tipo)}mll', ha='center', va='center', fontsize=15, fontweight='bold')

    # ax[1].set_title(f'Ventas por {variable_cualitatitva}', fontsize=25, color='darkblue', fontweight='bold', pad=30)
    ax[1].margins(.1)

    plt.suptitle(f'{variable_cuantitativa} por {variable_cualitatitva}', fontsize=25, fontweight='bold',color='darkblue')

    plt.show()
```

```python
#Creamos nuestra funcion que determinara el comportamiento de numero de clusteres
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score

def calcular_inercia_clusteres(datos):

  inercia = []
  silhouette_scores = []
  davies_bouldin_scores = []
  calinski_harabasz_scores = []

  for k in range(2,11):

    kmeans = KMeans(n_clusters=k)

    kmeans.fit(X = datos)

    train_cluster_labels = kmeans.fit_predict(datos)

    # Calcular el coeficiente de Silhouette para los datos de entrenamiento
    train_silhouette_avg = silhouette_score(datos, train_cluster_labels)

    # Calcular el Ã­ndice Davies-Bouldin para los datos de entrenamiento
    train_davies_bouldin = davies_bouldin_score(datos, train_cluster_labels)

    # Calcular el Ã­ndice Calinski-Harabasz para los datos de entrenamiento
    train_calinski_harabasz = calinski_harabasz_score(datos, train_cluster_labels)


    #Almacenamos el valor de silhoulette en un array
    silhouette_scores.append([k, train_silhouette_avg])

    #Almacenamos el valor de DAvies-BOuldin en un array
    davies_bouldin_scores.append([k, train_davies_bouldin])

    #Almacenamos el valor de calinski_harabasz en un array
    calinski_harabasz_scores.append([k,train_calinski_harabasz])

    #Cargamos los valores de la inercia al arreglo inercia
    inercia.append(kmeans.inertia_)


  return [inercia,silhouette_scores, davies_bouldin_scores, calinski_harabasz_scores]
```

```python
#Dataframes de aspectos
x_ber = df_selected.drop(columns='miembro', axis = 1)
y_ber = df_selected['miembro']

#Generamos el modelo de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(
    x_ber,
    y_ber,
    test_size = 0.25,
    random_state = 123
)

#Calculamos el dataframe de la inercia
df_inercia = calcular_inercia_clusteres(X_train)
```
